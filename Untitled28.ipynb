```python
import pandas as pd

# read the csv file

df = pd.read_csv('netflix_titles.csv')
df.head(10)
```

```python
# shape of the dataframe
df.shape
```

```python
# data type of each column
df.dtypes
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# histogram of the release_year
df.release_year.plot(kind='hist', color='purple', edgecolor='black', figsize=(10,7))
plt.title('Distribution of release_year', size=24)
plt.xlabel('release_year (inches)', size=18)
plt.ylabel('Frequency', size=18)

# histogram of the ratingof10
df.ratingof10.plot(kind='hist', color='purple', edgecolor='black', figsize=(10,7))
plt.title('Distribution of ratingof10', size=24)
plt.xlabel('ratingof10 (pounds)', size=18)
plt.ylabel('Frequency', size=18);
```

```python

# histogram of the release_year - movies and tv shows

df[df['type'] == 'Movie'].release_year.plot(kind='hist', color='blue', edgecolor='black', alpha=0.5, figsize=(10, 7))
df[df['type'] == 'TV Show'].release_year.plot(kind='hist', color='magenta', edgecolor='black', alpha=0.5, figsize=(10, 7))
plt.legend(labels=['Movie', 'TV Show'])
plt.title('Distribution of release_year', size=24)
plt.xlabel('release_year (inches)', size=18)
plt.ylabel('Frequency', size=18);
```

```python
# histogram of the ratingof10 - movies and tv shows

df[df['type'] == 'Movie'].ratingof10.plot(kind='hist', color='blue', edgecolor='black', alpha=0.5, figsize=(10, 7))
df[df['type'] == 'TV Show'].ratingof10.plot(kind='hist', color='magenta', edgecolor='black', alpha=0.5, figsize=(10, 7))
plt.legend(labels=['Movie', 'TV Show'])
plt.title('Distribution of ratingof10', size=24)
plt.xlabel('ratingof10 (pounds)', size=18)
plt.ylabel('Frequency', size=18);
```

```python
# Descriptive statistics movie
statistics_movie = df[df['type'] == 'Movie'].describe()
statistics_movie.rename(columns=lambda x: x + '_movie', inplace=True)

# Descriptive statistics tv show
statistics_tvshow = df[df['type'] == 'TV Show'].describe()
statistics_tvshow.rename(columns=lambda x: x + '_tvshow', inplace=True)

# Dataframe that contains statistics for both movie and tv show
statistics = pd.concat([statistics_movie, statistics_tvshow], axis=1)
statistics
```

```python
# Scatter plot of release_year and ratingof10

ax1 = df[df['type'] == 'Movie'].plot(kind='scatter', x='release_year', y='ratingof10', color='blue', alpha=0.5, figsize=(10, 7))
df[df['type'] == 'TV Show'].plot(kind='scatter', x='release_year', y='ratingof10', color='magenta', alpha=0.5, figsize=(10 ,7), ax=ax1)
plt.legend(labels=['Movies', 'TV Shows'])
plt.title('Relationship between release_year and ratingof10', size=24)
plt.xlabel('release_year (inches)', size=18)
plt.ylabel('ratingof10 (pounds)', size=18);
```

```python
# Scatter plot of 500 tv shows

sample_tvshows = df[df['type'] == 'TV Show'].sample(500)
sample_tvshows.plot(kind='scatter', x='release_year', y='ratingof10', color='magenta', alpha=0.5, figsize=(10, 7))
plt.legend(labels=['TV Show'])
plt.title('Relationship between release_year and ratingof10 (sample of 500 tvshows)', size=20)
plt.xlabel('release_year (inches)', size=18)
plt.ylabel('ratingof10 (pounds)', size=18);
```

```python
import numpy as np

# best fit polynomials

df_movies = df[df['type'] == 'Movie']
df_tvshows = df[df['type'] == 'TV Show']

# polynomial - movies
movie_fit = np.polyfit(df_movies.release_year, df_movies.ratingof10, 1)
# array([   5.96177381, -224.49884071])

# polynomial - tv shows
tvshow_fit = np.polyfit(df_tvshows.release_year, df_tvshows.ratingof10, 1)
# array([   5.99404661, -246.01326575])
```

```python
# scatter plots and regression lines

# movies and tv shows dataframes
df_movies = df[df['type'] == 'Movie']
df_tvshows = df[df['type'] == 'TV Show']

# Scatter plots.
ax1 = df_movies.plot(kind='scatter', x='release_year', y='ratingof10', color='blue', alpha=0.5, figsize=(10, 7))
df_tvshows.plot(kind='scatter', x='release_year', y='ratingof10', color='magenta', alpha=0.5, figsize=(10, 7), ax=ax1)

# regression lines
plt.plot(df_movies.release_year, movie_fit[0] * df_movies.release_year + movie_fit[1], color='darkblue', linewidth=2)
plt.plot(df_tvshows.release_year, tvshow_fit[0] * df_tvshows.release_year + tvshow_fit[1], color='deeppink', linewidth=2)

# regression equations
plt.text(65, 230, 'y={:.2f}+{:.2f}*x'.format(movie_fit[1], movie_fit[0]), color='darkblue', size=12)
plt.text(70, 130, 'y={:.2f}+{:.2f}*x'.format(tvshow_fit[1], tvshow_fit[0]), color='deeppink', size=12)

# legend, title and labels.
plt.legend(labels=['Movies Regresion Line', 'TV Show Regresion Line', 'Movies', 'TV Show'])
plt.title('Relationship between release_year and ratingof10', size=24)
plt.xlabel('release_year (inches)', size=18)
plt.ylabel('ratingof10 (pounds)', size=18);
```

```python
import seaborn as sns

# regression plot using seaborn
fig = plt.figure(figsize=(10, 7))
sns.regplot(x=df_movies.release_year, y=df_movies.ratingof10, color='blue', marker='+')
sns.regplot(x=df_tvshows.release_year, y=df_tvshows.ratingof10, color='magenta', marker='+')

# Legend, title and labels.
plt.legend(labels=['Movies', 'TV Shows'])
plt.title('Relationship between release_year and ratingof10', size=24)
plt.xlabel('release_year (inches)', size=18)
plt.ylabel('ratingof10 (pounds)', size=18);
```
